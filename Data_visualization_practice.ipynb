```python
x = linspace(0, 5, 10)
y = x ** 2
print(x)
print(y)
```

```python
figure()
plot(x,y,'r')
xlabel('x')
ylabel('y')
title('title')
show()
```

```python
%matplotlib inline
```

```python
figure()
plot(x,y,'r')
xlabel('x')
ylabel('y')
title('title')
show()
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

axes.plot(x,y,'r')

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # left, bottom, width, height (range 0 to 1)

axes1.plot(x,y,'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('title')

axes2.plot(y,x,'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('insert title');
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)

for ax, color in zip(axes, ['r', 'g']):
    ax.plot(x,y,color)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)

for ax, color in zip(axes, ['r', 'g']):
    ax.plot(x,y,color)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
fig.tight_layout()
```

```python
fig, axes = plt.subplots(figsize=(12,3))

axes.plot(x,y,'b')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

```python
fig, ax = plt.subplots()

ax.plot(x,x**2,label="y = x**2")
ax.plot(x,x**3,label="y = x**3")
ax.legend(loc=2) # upper left corner
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title');
```

```python
from numpy import random
n = random.randn(100000)
fig, ax = plt.subplots(1,2,figsize=(12,4))

ax[0].hist(n)
ax[0].set_title('hist')
ax[0].set_xlim((min(n), max(n)))

ax[1].hist(n, cumulative=True, bins=50)
ax[1].set_title('cumulative hist')
ax[1].set_xlim((min(n), max(n)));
```
